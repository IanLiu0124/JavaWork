## Importing the neccessary files 

```python
#Load the python library!
import numpy as np
import pandas as pd
```

## Reading the data

```python
#Import and read Data
df = pd.read_csv('dataset_002.csv')
df.head()
```

## Displaying the summary of the data type

```python
#data type summary
print('\n\n')
print(df.dtypes)
```

## Dropping and Filling Data: proper naming.

```python
# number of null
null_counts = df.isnull().sum()
print(null_counts)
#Dropping columns and rows where all values are null
df.dropna(how='all', inplace=True)
df.dropna(axis= 1, how='all', inplace = True)
```

```python
df.drop(columns=['id', 'URL', 'REGION', 'region URL', 'VIN', 'LATITUDE', 'LONGITUDE', 'DATE'], inplace=True)
#dropping unneeded columns
```

```python
odometerNull= df['odometer'].isnull().sum()
milesNull = df['miles'].isnull().sum()

if odometerNull > milesNull:
    df.drop(columns=['odometer'], inplace = True)
elif milesNull > odometerNull:
    df.drop(columns=['miles'], inplace = True)
    
df.head()
```

## Dropping Rows Using Threshold of Empty Cells

```python
#Chceck number of null for each column. Drop with threshhold of 10
print(null_counts)
df.dropna(thresh=df.shape[1] - 10, inplace=True)
# df.head() 
```

## Changing Columns Label

```python
#Replace the whitespaces into underscore “_” and change all letters into small case letters. 
#You can use the str.replace() to the df.columns or use a lambda function to perform that task.
#Also replace the string __ (two underscore) into a single underscore in column manufacturer__model
df.columns = df.columns.str.replace(' ', '_').str.lower()

df.columns = df.columns.str.replace('__', '_')

# df.head()
```

## Working on the Manufacturer_model column

```python
#Working on the manufacturer model. To split the manufacturer from model

df['manufacturer'] = df['manufacturer_model'].str.split('_').str[0]
df.drop(columns=['manufacturer_model'], inplace = True)
df.dropna(subset=['manufacturer'], inplace = True)
df.head()
```

## Working on the price column

```python
#Remove character $ sign and , comma from Price Coluimn

df['price'] = df['price'].str.replace('$', '').str.replace(',', '')
#converting the data to numeric
df['price'] = pd.to_numeric(df['price'])
#dropping nule values
nan_indices = df[df['price'].isna()].index
df.drop(index=nan_indices, inplace=True)
#creating cad column
df['price_cad'] = 1.36 * df['price']
#lastly, dropping the column
df.drop(columns=['price'], inplace=True)
df.head()
```

## Reducing further Nan rows

```python
#Reducing further NaN rows
#empty year, miles, fuel and status values are all to be dropped

print(df[['year', 'miles', 'fuel', 'status']].isnull().sum())
df.dropna(subset=['year', 'miles', 'fuel', 'status'], inplace=True)

print(df[['year', 'miles', 'fuel', 'status']].isnull().sum())
```

## Workingg on the color column

```python
#check the number of null in t he column color
print(df['color'].isnull().sum())
#find index of all nan and fill it with unknown

df['color'] = df['color'].fillna('unknown')
print(df['color'].isnull().sum())

#use value_counts() to check different car colours in dataset

color = df['color'].value_counts()

colorLess =color[color < 500].index

df.loc[df['color'].isin(colorLess), 'color'] = 'other'

print(df['color'].value_counts())
```

## Working on the condition column

```python
#column condition. Assume that any car data that does not have any info about its condition will have unknown value. Find and fill with unknown

df['condition']=df['condition'].fillna('unknown')
print(df['condition'].value_counts())
```

## Working on the cylinders column

```python
#Cylinders column has a combination of number and text. Use value_counts() to see that the cylinders column has either in the form of X cylinders or others
#Assume most cars have 4, find all index and fill with 4 cylinders
print(df['cylinders'].value_counts())
df['cylinders'] = df['cylinders'].fillna('4 cylinders')
#remove the text 'cylinders' from the values
df['cylinders'] = df['cylinders'].str.replace('cylinders', '')
#drop rows that have 'others' in the cylinder
df = df[df['cylinders'] != 'other']
df.head()
```

## Remove the remaining NaN and save the cleaned File

```python
#Use dropna function to drop remaining 

df.dropna(inplace = True)
df.reset_index(drop=True, inplace=True)
df.to_csv('Lab1_cleaned_ILi688.csv', index=False)
```
