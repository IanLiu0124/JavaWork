```python
df_filtered = df[df['price_cad'] <= 100000]
df_filtered['price_cad'] = pd.to_numeric(df_filtered['price_cad'], errors='coerce')
df_filtered.loc[:, 'cbrt_price_cad'] = np.cbrt(df_filtered['price_cad'])
df_filtered.drop('price_cad', axis=1, inplace=True)
df_filtered = df_filtered[df_filtered['cbrt_price_cad'] > 10]
```

```python

# Distribution plot of year
sns.histplot(df_filtered['year'], kde=True)
plt.title('Distribution of Year')
plt.show()

# Dropping records where year <= 1980
df_filtered = df_filtered[df_filtered['year'] > 1980]

# Distribution plot of year again
sns.histplot(df_filtered['year'], kde=True)
plt.title('Distribution of Year After Dropping Records')
plt.show()

# Countplot for status and type
sns.countplot(data=df_filtered, x='status')
plt.title('Count of Status')
plt.show()

sns.countplot(data=df_filtered, x='type')
plt.title('Count of Type')
plt.show()
```

```python
# cbrt_price_cad against miles
sns.scatterplot(data=df_filtered, x='miles', y='cbrt_price_cad')
plt.title('cbrt_price_cad vs Miles')
plt.show()

# Dropping records where miles >= 500000
df_filtered = df_filtered[df_filtered['miles'] < 500000]

# cbrt_price_cad against year
sns.scatterplot(data=df_filtered, x='year', y='cbrt_price_cad')
plt.title('cbrt_price_cad vs Year')
plt.show()

# cbrt_price_cad against type
sns.boxplot(data=df_filtered, x='type', y='cbrt_price_cad')
plt.title('cbrt_price_cad vs Type')
plt.show()
```

```python
print(df_filtered['condition'].value_counts())
print(df_filtered['fuel'].value_counts())
print(df_filtered['status'].value_counts())
print(df_filtered['transmission'].value_counts())
print(df_filtered['drive'].value_counts())
print(df_filtered['type'].value_counts())
print(df_filtered['color'].value_counts())
print(df_filtered['manufacturer'].value_counts())

# Dropping records with condition = 'salvage'
df_filtered = df_filtered[df_filtered['condition'] != 'salvage']

# Dropping records with missing status
df_filtered = df_filtered[df_filtered['status'].notnull()]

# Average value of value_counts for manufacturers
m_mean = df_filtered['manufacturer'].value_counts().mean()
filter = df_filtered['manufacturer'].value_counts() < m_mean
m_list_other = filter[filter==True].index.to_list()

# Changing manufacturers with low counts to 'other'
df_filtered['manufacturer'] = df_filtered['manufacturer'].replace(m_list_other, 'other')

# Replacing hyphens and spaces in categorical columns
categorical_columns = ['condition', 'fuel', 'status', 'transmission', 'drive', 'type', 'color', 'manufacturer']
for col in categorical_columns:
    df_filtered[col] = df_filtered[col].str.replace('-', '_').str.replace(' ', '_').str.lower()
```

```python
# Generating dummy variables for categorical columns
fu_dummy = pd.get_dummies(df_filtered['fuel'], drop_first=True, prefix='fu')
df_filtered = df_filtered.join(fu_dummy)
df_filtered.drop('fuel', axis=1, inplace=True)

st_dummy = pd.get_dummies(df_filtered['status'], drop_first=True, prefix='st')
df_filtered = df_filtered.join(st_dummy)
df_filtered.drop('status', axis=1, inplace=True)

tr_dummy = pd.get_dummies(df_filtered['transmission'], drop_first=True, prefix='tr')
df_filtered = df_filtered.join(tr_dummy)
df_filtered.drop('transmission', axis=1, inplace=True)

dr_dummy = pd.get_dummies(df_filtered['drive'], drop_first=True, prefix='dr')
df_filtered = df_filtered.join(dr_dummy)
df_filtered.drop('drive', axis=1, inplace=True)

ty_dummy = pd.get_dummies(df_filtered['type'], drop_first=True, prefix='ty')
df_filtered = df_filtered.join(ty_dummy)
df_filtered.drop('type', axis=1, inplace=True)

co_dummy = pd.get_dummies(df_filtered['condition'], prefix='co')
df_filtered = df_filtered.join(co_dummy)
df_filtered.drop('co_unknown', axis=1, inplace=True)  # Assuming you have an unknown category
df_filtered.drop('condition', axis=1, inplace=True)

cl_dummy = pd.get_dummies(df_filtered['color'], prefix='cl')
df_filtered = df_filtered.join(cl_dummy)
df_filtered.drop('cl_unknown', axis=1, inplace=True)  # Assuming you have an unknown category
df_filtered.drop('color', axis=1, inplace=True)

mk_dummy = pd.get_dummies(df_filtered['manufacturer'], prefix='mk')
df_filtered = df_filtered.join(mk_dummy)
df_filtered.drop('mk_other', axis=1, inplace=True)  # Assuming you have an 'other' category
df_filtered.drop('manufacturer', axis=1, inplace=True)

# Checking the shape of the dataframe
print(df_filtered.shape) 
```

```python
# Displaying the correlation to cbrt_price_cad
correlation_cbrt = df_filtered.corr()['cbrt_price_cad'].abs().sort_values(ascending=False)
print(correlation_cbrt)
```
